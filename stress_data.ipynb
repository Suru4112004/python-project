```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/corporate_stress_dataset.csv')

# 1. Gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Gender', palette='Set2')
plt.title('Gender Distribution')
plt.show()
```

```python
# 2. Stress level vs. working hours per week
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Working_Hours_per_Week', y='Stress_Level', hue='Gender', alpha=0.7)
plt.title('Stress Level vs. Working Hours per Week')
plt.show()
```

```python
# 3. Experience years vs. monthly salary
plt.figure(figsize=(8, 6))
sns.regplot(data=data, x='Experience_Years', y='Monthly_Salary_INR', color='blue', scatter_kws={"alpha": 0.5})
plt.title('Experience Years vs. Monthly Salary')
plt.show()
```

```python
# 4. Work-life balance by job roles
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Job_Role', y='Work_Life_Balance', palette='coolwarm')
plt.xticks(rotation=45, ha='right')
plt.title('Work-Life Balance by Job Roles')
plt.show()
```

```python
# 5. Remote work vs. stress level
plt.figure(figsize=(6, 4))
sns.boxplot(data=data, x='Remote_Work', y='Stress_Level', palette='Set3')
plt.title('Remote Work vs. Stress Level')
plt.show()

```

```python

# 6. Physical activity vs. burnout symptoms
plt.figure(figsize=(8, 4))
sns.violinplot(data=data, x='Burnout_Symptoms', y='Physical_Activity_Hours_per_Week', palette='muted')
plt.title('Physical Activity vs. Burnout Symptoms')
plt.show()

```

```python
# 7. Sleep hours vs. stress level
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Sleep_Hours', y='Stress_Level', hue='Burnout_Symptoms', alpha=0.7)
plt.title('Sleep Hours vs. Stress Level')
plt.show()
```

```python
# 8. Gender bias and job satisfaction
plt.figure(figsize=(6, 4))
sns.boxplot(data=data, x='Gender_Bias_Experienced', y='Job_Satisfaction', palette='cool')
plt.title('Gender Bias vs. Job Satisfaction')
plt.show()
```

```python
# 9. Commute time vs. work pressure level
plt.figure(figsize=(8, 6))
sns.regplot(data=data, x='Commute_Time_Hours', y='Work_Pressure_Level', scatter_kws={"alpha": 0.5}, color='green')
plt.title('Commute Time vs. Work Pressure Level')
plt.show()
```

```python
# 10. Team size vs. performance rating
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Team_Size', y='Performance_Rating', alpha=0.6)
plt.title('Team Size vs. Performance Rating')
plt.show()
```

```python
# Conclusions
print("\nOverall Conclusions:")
print("1. High working hours, long commutes, and insufficient sleep are significant drivers of workplace stress.")
print("2. Gender bias negatively affects job satisfaction, highlighting the need for inclusive policies.")
print("3. Work-life balance varies significantly across job roles, and interventions can be role-specific.")
print("4. Remote work and physical activity positively contribute to reduced stress and burnout symptoms.")
print("5. Experience is rewarded monetarily, but optimal team sizes and shorter commutes can enhance performance and satisfaction.")
```
